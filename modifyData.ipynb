```python
#creating bins of the same size
num_bins = 8

def split(a, n):
    k, m = divmod(len(a), n)
    return (a[i*k+min(i, m):(i+1)*k+min(i+1, m)] for i in range(n))

pops = trails["popularity"].sort_values().to_list()
bins = list(split(pops, num_bins))

for idx, b in enumerate(bins):
    print("bin: {}, binsize: {}".format(idx + 1, len(b)))
    print("bin average: {}".format(np.mean(b)))
    print("max val: {}, min val: {}".format(max(b), min(b)))
    print("standard deviation: {}\n".format(stat.pstdev(b)))
    
```

```python
#creating bins of the same size - adjusted to account for largest values
num_bins = 8

def split(a, n):
    k, m = divmod(len(a), n)
    return (a[i*k+min(i, m):(i+1)*k+min(i+1, m)] for i in range(n))

pops = trails["popularity"].sort_values().to_list()
pops_upper = [p for p in pops if p > 35]
pops_minus_upper = sorted(list(set(pops) - set(pops_upper)))
bins = list(split(pops_minus_upper, num_bins))

for idx, b in enumerate(bins):
    print("bin: {}, binsize: {}".format(idx + 1, len(b)))
    print("bin average: {}".format(np.mean(b)))
    print("max val: {}, min val: {}".format(max(b), min(b)))
    print("standard deviation: {}\n".format(stat.pstdev(b)))
    
print("bin: {}, binsize: {}".format(num_bins + 1, len(pops_upper)))
print("bin average: {}".format(np.mean(pops_upper)))
print("max val: {}, min val: {}".format(max(pops_upper), min(pops_upper)))
print("standard deviation: {}\n".format(stat.pstdev(pops_upper)))
```
